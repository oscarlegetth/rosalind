```python
protein_file = open("protein.txt", "r")
protein = {}
for line in protein_file:
    words = line.split(" ")
    protein[words[0]] = words[1][:-1].upper()
protein_file.close()
# print(protein)
input_file = open("rosalind_prot.txt", "r")
protein_string = []
line = input_file.readline()[:-1]
buffer = ""
for char in line:
    buffer += char
    if (len(buffer) == 3):
        protein_string.append(protein[buffer])
        buffer = ""
input_file.close()
print(''.join(protein_string[:-1]))
```
