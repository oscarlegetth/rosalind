```python
file = open("rosalind_grph.txt", "r")
suffixes = {}
prefixes = {}
while True:
    rosalind_id = file.readline().strip()[1:]
    if not rosalind_id:
        break
    dna = ""
    while True:
        dna += file.readline().strip()
        if not dna:
            break
        pos = file.tell()
        next_line = file.readline()[:1]
        file.seek(pos)
        if (not next_line or next_line == ">"):
            break
    
    prefix = dna[:3]
    suffix = dna[-3:]
    if (prefix not in prefixes):
        prefixes[prefix] = []
    if (suffix not in suffixes):
        suffixes[suffix] = []
    prefixes[dna[:3]].append(rosalind_id)
    suffixes[dna[-3:]].append(rosalind_id)

for key in prefixes:
    if key in suffixes:
        ids_1, ids_2 = suffixes[key], prefixes[key]
        for s in ids_1:
            for p in ids_2:
                if p != s:
                    print(s, p)
file.close()
```
