```python
file = open("rosalind_lcsm.txt", "r")
longest = ""
dna_list = []
while True:
    rosalind_id = file.readline().strip()[1:]
    if not rosalind_id:
        break
    dna = ""
    while True:
        dna += file.readline().strip()
        if not dna:
            break
        pos = file.tell()
        next_line = file.readline()[:1]
        file.seek(pos)
        if (not next_line or next_line == ">"):
            break
    dna_list.append(dna)
    if (len(dna) > len(longest)):
        longest = dna
   
file.close()

for i in range(len(longest), 0, -1):
    for j in range(0, len(longest) - i + 1):
        substring = longest[j:i+j]
        common_substring_found = True
        for s in dna_list:
            if substring not in s:
                common_substring_found = False
                break
        if common_substring_found:
            print(substring)
            
        
```
